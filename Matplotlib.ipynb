#Bar


```python
Fav_food = {"pizza":3,
            'pasta':5,
            'meat':7}
```

```python
fig, ax = plt.subplots()
ax.bar(Fav_food.keys(),Fav_food.values())
## set > title and labelling
ax.set(title= 'Italian_Food', xlabel ="food",ylabel = 'price')
```

#BARH > HORIZONTAL BARCHART

```python
fig, ax = plt.subplots()
ax.barh(list(Fav_food.keys()),list(Fav_food.values()))
## set > title and labelling
ax.set(title= 'Italian_Food', xlabel ="price",ylabel = 'food')
```

```python
#34-Create a histogram of the age column with 50 bins

plt.hist(df2['age'], bins = 50)
```

```python
#36
df2.plot.hist(subplots = True, figsize = (10,30))
```

```python
df2 = df2[df2['age']>50]
df2
```

```python
#37-*accounts for unlimited number of variable
fig, ax = plt.subplots(figsize = (10,10))
scatter = ax.scatter(df2['age'], df2['chol'], c = df2['target'])
ax.set (title = 'heart disease and cholesterol level', xlabel = 'Age',ylabel= 'Cholesterol')
ax.legend(*scatter.legend_elements(),title = 'Target')
ax.axhline(df2['chol'].mean(), linestyle = '--')

```

```python
#40
fig, ax = plt.subplots(figsize = (10,10))
scatter = ax.scatter(df2['age'], df2['chol'], c = df2['target'], cmap= 'winter')
ax.set (title = 'heart disease and cholesterol level', xlabel = 'Age',ylabel= 'Cholesterol')
ax.legend(*scatter.legend_elements(),title = 'Target')
ax.axhline(df2['chol'].mean(), linestyle = '--', color = 'red')
```

```python
#Build a plot with 2 subplots

fig, (ax0, ax1) = plt.subplots(nrows = 2, ncols = 1,figsize=(10,10), sharex = True)
scatter = ax0.scatter(df2['age'], df2['chol'], c = df2['target'], cmap= 'winter')
ax0.set (title = 'heart disease and cholesterol levels', ylabel= 'Cholesterol')
ax0.legend(*scatter.legend_elements(),title = 'Target')
ax0.axhline(df2['chol'].mean(), linestyle = '--', color = 'blue')



scatter = ax1.scatter(df2['age'], df2['thalach'], c = df2['target'], cmap= 'winter')
ax1.set (title = 'heart disease and Max Heart Rate Level',xlabel = 'Age', ylabel= 'Max Heart Rate')
ax1.legend(*scatter.legend_elements(),title = 'Target')
ax1.axhline(df2['thalach'].mean(), linestyle = '--', color = 'red')
fig.suptitle('Heart Disease Analysis', fontsize = 15, fontweight = 'bold')
```
